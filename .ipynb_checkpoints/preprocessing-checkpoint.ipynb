```python
ingredients = recipes.columns.tolist()[1:] # take all values starting from first index instead of 0
```

```python
import re
```

```python
for ingredient in ingredients:
    match = re.compile(".*(sauce).*").search(ingredient)
    if match:
        print(match.group(0))
```

```python
plt.figure(figsize=(15,5))
sns.countplot(x='country',data=recipes)
plt.xticks(rotation=90)
plt.show()
```

```python
recipes_counts = recipes["country"].value_counts()
cuisines_indices = recipes_counts > 50

cuisines_to_keep = list(np.array(recipes_counts.index.values)[np.array(cuisines_indices)])
print(cuisines_to_keep)
```

```python
recipes = recipes.loc[recipes['country'].isin(cuisines_to_keep)]
```

```python
recipes = recipes.replace(to_replace="Yes", value=1)
recipes = recipes.replace(to_replace="No", value=0)
recipes
```

```python
for i in range(5):
    recipe1 = recipes.iloc[i].reset_index()
    print(recipe1[recipe1[i] == 1]['index'].tolist())
```
