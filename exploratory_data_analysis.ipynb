### example ML algo implementation

```python
sample_df = df[['horsepower','price']].copy()
x = df[['horsepower']]
y = df['price']
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(x,y)
```

```python
sample_df['predictions'] = model.predict(x)
```

```python
sample_df
```

```python
ax = sample_df.plot(x='horsepower',y='price',style='o')
sample_df.plot(x='horsepower',y='predictions',style='-',ax=ax)
```
